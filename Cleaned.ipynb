# Dropping Columns

At least four of the colums in the dataframe are not needed for modeling. "step", "nameDest", "isFlaggedFraud" and "nameOrig" don't offer much information that will make out models more accurate and will be dropped.

```python
# Drop the columns that don't make our model more accurate
columns_to_drop = ["step", "nameDest", "nameOrig", "isFlaggedFraud"]
accurate_df = raw_df.drop(columns= columns_to_drop, axis= 1)
#drop any possible duplicates
accurate_df = accurate_df.drop_duplicates()
accurate_df
```

```python
# analyze the shape of the  cleaned data
accurate_df.shape
```

```python
# Use one-hot encoding on the type column to convert to int

cleaned_df = pd.get_dummies(accurate_df, columns=['type'], dtype=int, drop_first=True)
```
